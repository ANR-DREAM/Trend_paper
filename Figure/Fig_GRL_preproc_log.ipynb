```python
#Figure 3  - January 2024
# same for log(chl)
#Preprocessing of Chl-a inputs into monthly median
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import pandas as pd
import glob
import dask

from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels
import statsmodels.api as sm

import cmocean
import colorcet as cc
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
cartopy.config["data_dir"] = './cartopy_shapefiles'

ds_input = xr.open_mfdataset('/home/datawork-lops-oh/biogeo/AI/CNN_CHLORO/INPUT/Input_obs9_*.nc')
ds_input = ds_input.transpose('time', 'latitude', 'longitude')
a = ds_input.mask.where(ds_input.mask==6)
mask = np.isnan(a)
mask = ~mask   
ds_input = ds_input.assign(variables={"mask": (('latitude','longitude'), mask.data)}) 
```

```python
#CCI
ds_out = xr.open_mfdataset('/home2/datawork/epauthen/Ocean-Colour-CCI/OC_CCI_Coarse/OC_CCI_chloro_a_*.nc')
ds_out = ds_out.rename({'chlor_a_coarse':'chloro'})
ds_out = ds_out.sel(latitude = slice(-50,50))
ds_out = ds_out.assign(variables={"mask": (('latitude','longitude'), ds_input.mask.data)}) 
ds_out = ds_out.assign(variables={"chl_log": (('time','latitude','longitude'), np.log(ds_out.chloro.data))})
y_test = ds_out.where(ds_out.mask == 1).load()
#
# Compute weighted median by timestep
weights = np.cos(np.deg2rad(y_test.latitude))
weights.name = "weights"
y_testw = y_test.chl_log.weighted(weights)
chloro_qua = y_testw.quantile(dim = ("longitude", "latitude"),q=0.5)
CCI = y_test.assign(variables={"chl_log_qua": (('time'), chloro_qua.data)}) 
```

```python
#Load CCI pred ensemble mean
with dask.config.set(**{'array.slicing.split_large_chunks': False}):
    ds_out = xr.open_dataset('/home/datawork-lops-oh/biogeo/AI/CNN_CHLORO/OUTPUT/Model_GRL/CCI_ensemble_mean.nc')
ds_out = ds_out.rename({'chloro_pred': 'chl'})
ds_out = ds_out.assign(variables={"chl_log": (('time','latitude','longitude'), np.log(ds_out.chl.data))})
y_test = ds_out.where(ds_out.mask == 1).load()
#
# Compute weighted median by timestep
weights = np.cos(np.deg2rad(y_test.latitude))
weights.name = "weights"
y_testw = y_test.chl_log.weighted(weights)
chloro_qua = y_testw.quantile(dim = ("longitude", "latitude"),q=0.5)
y_pred = y_test.assign(variables={"chl_pred_log_qua": (('time'), chloro_qua.data)}) 
```

```python
fig = plt.subplots(1,1, figsize=(10,3), dpi=300)
CCI.chl_log_qua.plot(label = "log(chl) CCI",color = "black")
y_pred.chl_pred_log_qua.plot(label = "log(chl) CCI pred",color = "tab:red")
d1 = '2002-07'
d2 = '2009-12'
plt.axvline(x = pd.to_datetime(d1),linestyle = '--',color = 'black')
plt.axvline(x = pd.to_datetime(d2),linestyle = '--',color = 'black')    
plt.grid()
plt.legend()
```

```python
#GSM
data_input = '/home2/datawork/epauthen/Globcolour_coarse/'
with dask.config.set(**{'array.slicing.split_large_chunks': False}):
    MODVIR    = xr.open_mfdataset(data_input + "MODVIR_*.nc")
    MERMOD    = xr.open_mfdataset(data_input + "MERMOD_*.nc")
    MERMODVIR = xr.open_mfdataset(data_input + "MERMODVIR_*.nc")
    MERMODSWF = xr.open_mfdataset(data_input + "MERMODSWF_*.nc")
    SWF    = xr.open_mfdataset(data_input + "SWF_*.nc")
    SWF    = SWF.sel(time = slice('1997-09','2002-06'))
    ds_gsm = MODVIR.merge(MERMOD)
    ds_gsm = ds_gsm.merge(MERMODSWF)
    ds_gsm = ds_gsm.merge(SWF)
    ds_out = ds_gsm.merge(MERMODVIR)
    ds_out = ds_out.rename({'CHL1_coarse': 'chloro'})
    ds_out = ds_out.sel(latitude = slice(-50,50))
    ds_out = ds_out.assign(variables={"mask": (('latitude','longitude'), ds_input.mask.data)}) 
    ds_out = ds_out.assign(variables={"chl_log": (('time','latitude','longitude'), np.log(ds_out.chloro.data))})
    y_test = ds_out.where(ds_out.mask == 1).load()
#
# Compute weighted median by timestep
weights = np.cos(np.deg2rad(y_test.latitude))
weights.name = "weights"
y_testw = y_test.chl_log.weighted(weights)
chloro_qua = y_testw.quantile(dim = ("longitude", "latitude"),q=0.5)
GSM = y_test.assign(variables={"chl_log_qua": (('time'), chloro_qua.data)}) 
```

```python
#Load GSM pred ensemble mean
with dask.config.set(**{'array.slicing.split_large_chunks': False}):
    ds_out = xr.open_dataset('/home/datawork-lops-oh/biogeo/AI/CNN_CHLORO/OUTPUT/Model_GRL/GSM_ensemble_mean.nc')
ds_out = ds_out.rename({'chloro_pred': 'chl'})
ds_out = ds_out.assign(variables={"chl_log": (('time','latitude','longitude'), np.log(ds_out.chl.data))})
y_test = ds_out.where(ds_out.mask == 1).load()
#
# Compute weighted median by timestep
weights = np.cos(np.deg2rad(y_test.latitude))
weights.name = "weights"
y_testw = y_test.chl_log.weighted(weights)
chloro_qua = y_testw.quantile(dim = ("longitude", "latitude"),q=0.5)
GSM_pred = y_test.assign(variables={"chl_pred_log_qua": (('time'), chloro_qua.data)}) 
```

```python
fig = plt.subplots(1,1, figsize=(10,3), dpi=300)
GSM.chl_log_qua.plot(label = "log(chl) GSM",color = "black")
GSM_pred.chl_pred_log_qua.plot(label = "log(chl) GSM pred",color = "tab:red") 
d1 = '2002-07'
d2 = '2009-12'
plt.axvline(x = pd.to_datetime(d1),linestyle = '--',color = 'black')
plt.axvline(x = pd.to_datetime(d2),linestyle = '--',color = 'black')    
plt.grid()
plt.legend()
```
